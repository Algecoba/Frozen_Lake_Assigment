```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# Cargar datos
df = pd.read_csv("datos_experimentales_completos.csv")

# Convertir Penalizacion a categ√≥rica
df['Penalizacion'] = df['Penalizacion'].astype('category')

# Modelo ANOVA factorial
model = ols(
    'Exito_Final ~ C(Penalizacion) * Alpha * Gamma',
    data=df
).fit()

anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

```python
anova_table['eta_sq'] = (
    anova_table['sum_sq'] /
    anova_table['sum_sq'].sum()
)

anova_table
```
